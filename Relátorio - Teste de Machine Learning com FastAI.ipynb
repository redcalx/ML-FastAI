# Relátorio - Teste de Machine Learning com FastAI

Neste relatório irei especificar as etapas executadas no teste de Machine Learning com FastAI e um modelo de Resnet18, ensinado na matéria de Inteligência Artificial do curso de Ciência da Computação da Universidade Federal do Amapá (UNIFAP), lecionado pelo Prof. Clay Palmeira. O modelo foi treinado para identificar 4 espécies diferentes de peixes amazônicos: Tucunaré, Tambaqui, Piraíba (Filhote) e Pirarucu. 

### 1ª Matriz de Confusão
